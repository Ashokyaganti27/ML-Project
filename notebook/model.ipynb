```python
import pandas as pd
```

```python
data=pd.read_csv("gemstone_trained.csv")
```

```python
data.head()
```

```python
x=data.drop(["price","Unnamed: 0"],axis=1)
```

```python
x.head()
```

```python
y=data["price"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
model=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.27,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
x_train=scalar.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=scalar.transform(x_test)
```

```python
x_test
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import ElasticNet
```

```python
estimators=[("lr",LinearRegression()),("dt",DecisionTreeRegressor(max_depth=8)),('elastic_net', ElasticNet(alpha=1.0, l1_ratio=0.5))]
```

```python
from sklearn.ensemble import StackingRegressor
```

```python
stacking_regressor=StackingRegressor(estimators=estimators,final_estimator=LinearRegression(),cv=5,passthrough=True,verbose=2)
```

```python
stacking_regressor.fit(x_train,y_train)
```

```python
stacking_regressor.score(x_train,y_train)
```

```python
y_pred_stackingregressor=stacking_regressor.predict(x_test)
```

```python
r2_score(y_test,y_pred_stackingregressor)
```

```python
from sklearn.ensemble import VotingRegressor
```
